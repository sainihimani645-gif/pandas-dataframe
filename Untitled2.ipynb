```python
# ==========================================================
# Assignment: Pandas DataFrame
# ==========================================================

import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Retail_dataset.csv')
df.head()

# ==========================================================
# Q1. Creating a DataFrame
# ==========================================================

data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Monitor', 'Keyboard'],
    'Region': ['East', 'West', 'North', 'South', 'East'],
    'Sales': [2500, 1800, 1200, 1500, 800],
    'Quantity': [5, 10, 8, 6, 15]
}

df = pd.DataFrame(data)

print("Q1: DataFrame")
print(df)

# ==========================================================
# Load Dataset for Questions 2–10
# ==========================================================

df = pd.read_csv("Retail_dataset.csv")

# ==========================================================
# Q2. Data Inspection
# ==========================================================

print("\nQ2: First 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())

print("\nDataset Info")
print(df.info())

print("\nDataset Shape")
print(df.shape)

# ==========================================================
# Q3. Column Selection
# ==========================================================

print("\nQ3: Product, Region, Sales Columns")
print(df[['Product', 'Region', 'Sales']])

# ==========================================================
# Q4. Indexing using .loc
# ==========================================================

print("\nQ4: Row with Index 10")
print(df.loc[10])

print("\nSales Value at Index 20")
print(df.loc[20, 'Sales'])

# ==========================================================
# Q5. Indexing using .iloc
# ==========================================================

print("\nQ5: First 6 Rows and First 4 Columns")
print(df.iloc[:6, :4])

print("\nSales Value from 3rd Row")
sales_col_index = df.columns.get_loc('Sales')
print(df.iloc[2, sales_col_index])

# ==========================================================
# Q6. Row Slicing
# ==========================================================

print("\nQ6: First 10 Rows")
print(df.head(10))

print("\nRows 15 to 25")
print(df.iloc[15:26])

# ==========================================================
# Q7. Filtering Data
# ==========================================================

print("\nQ7: Region = East")
print(df[df['Region'] == 'East'])

print("\nSales > 1500")
print(df[df['Sales'] > 1500])

# ==========================================================
# Q8. Multiple Condition Filtering
# ==========================================================

print("\nQ8: Region = West and Sales > 1200")
print(df[(df['Region'] == 'West') & (df['Sales'] > 1200)])

# ==========================================================
# Q9. Descriptive Statistics
# ==========================================================

print("\nQ9: Summary Statistics")
stats = df.describe()
print(stats)

print("\nMean Sales Value:")
print(df['Sales'].mean())

print("\nMaximum Quantity Sold:")
print(df['Quantity'].max())

print("\nMinimum Sales Value:")
print(df['Sales'].min())

# ==========================================================
# Q10. Aggregation Analysis
# ==========================================================

print("\nQ10: Aggregation Analysis")

print("Total Sales:", df['Sales'].sum())
print("Average Sales:", df['Sales'].mean())
print("Maximum Sales:", df['Sales'].max())
print("Minimum Sales:", df['Sales'].min())
```
